```python
%matplotlib inline

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
df1 = pd.read_csv('./data_ArduinoNano_parsed/data_parsed_1_deployed_stateOff.csv')
df2 = pd.read_csv('./data_ArduinoNano_parsed/data_parsed_2_deployed_stateStartUp.csv')

df1['datetime'] = pd.to_datetime(df1['datetime'])
df2['datetime'] = pd.to_datetime(df2['datetime'])

print(df1.shape, df2.shape)
print(df1.columns)

display(df1.head(3))
display(df2.head(3))
```

```python
col_compare = 'mag_y'

plt.figure(figsize=(20,15))
plt.style.use('seaborn-poster')
plt.subplot(2,1,1)
plt.plot(df1['datetime'], df1[col_compare])

plt.subplot(2,1,2)
plt.plot(df2['datetime'], df2[col_compare])
```

```python
plt.figure(figsize=(20,21))
plt.style.use('seaborn-poster')

plt.style.use('default')
plt.style.use('seaborn-whitegrid')
plt.style.use('seaborn-poster')

plt.subplot(3,1,1)
plt.plot(df1['datetime'], df1['mag_x'])
plt.ylabel('microTesla (X)')

plt.subplot(3,1,2)
plt.plot(df1['datetime'], df1['mag_y'])
plt.ylabel('microTesla (Y)')

plt.subplot(3,1,3)
plt.plot(df1['datetime'], df1['mag_z'])
plt.ylabel('microTesla (Z)')
plt.xlabel('Time');
```
